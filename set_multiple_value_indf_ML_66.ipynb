```python
# set multiple value in data frame
import pandas as pd
bond=pd.read_csv('jamesbond.csv',index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
mask=bond["Actor"] == "Roger Moore"
bond[mask]["Actor"] = "Rv"
```

```python
# using ix method we can use it to set multiple column values
mask = bond["Actor"] == "Sean Connery"
bond.ix[mask]
bond.ix[mask,"Actor"] = "Sir Sean Connery"
bond.head()
```

```python
# set multiple values usnig loc
mask = bond["Actor"] == "Daniel Craig"
bond[mask]
bond.loc[mask,"Actor"] = "Sir Daniel Craig"
bond[bond["Actor"] == "Sir Daniel Craig"]
```
