```python
# Pandas Rank method
import pandas as pd
nba=pd.read_csv('nba.csv').dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype(int)
nba
```

```python
nba["Salary"].rank()
# rank wise make salary
nba["Salary by Rank"] = nba["Salary"].rank(ascending=False).astype(int)
nba.sort_values("Salary",ascending=False)
```
