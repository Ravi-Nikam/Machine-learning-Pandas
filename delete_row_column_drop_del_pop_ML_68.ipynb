```python
# Delete row and columns in pandas
import pandas as pd
bond=pd.read_csv('jamesbond.csv',index_col="Film")
bond["Bond Actor Salary"].fillna(0,inplace=True)
bond
```

```python
# drop single value
bond.drop("Never Say Never Again")
```

```python
# drop multiple values
bond.drop(["The World Is Not Enough","Skyfall","Quantum of Solace"])
```

```python
# drop value column wise
bond.drop("Actor",axis=1)
```

```python
# drop multiple value column wise
bond.drop(["Actor","Director"],axis=1)
```

```python
# use of pop() to axtract values
actor=bond.pop("Actor")
bond
actor
```

```python
# use of del
del bond["Director"]
bond
```
