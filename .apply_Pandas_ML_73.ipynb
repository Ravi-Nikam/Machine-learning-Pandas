```python
# apply() in pandas
import pandas as pd
bond=pd.read_csv('jamesbond.csv')
bond.set_index("Film",inplace=True)
bond.sort_values("Film",inplace=True)
bond.head()
```

```python
def add_million(number):
    return str(number) + " Mill"

bond["Box Office"]=bond["Box Office"].apply(add_million)
bond.head()
```

```python
# add milion to multiple column
for col in ["Box Office","Budget","Bond Actor Salary"]:
    bond[col]=bond[col].apply(add_million)
    
bond.head()    
```
