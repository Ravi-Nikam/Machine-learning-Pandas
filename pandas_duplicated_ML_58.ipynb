```python
# duplicated() in pandas

import pandas as pd
df=pd.read_csv("employee.csv", parse_dates=['Start Date','Last Login Time'])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df.sort_values(["First Name"],inplace=True)
df.head()
```

```python
# it show first value as not duplicate and res duplicate by default keep = "First" so first is unique
df["First Name"].duplicated()
df[df["First Name"].duplicated()]
```

```python
# it  check from the last value the last first occur will be unique and res is duplicate
df["First Name"].duplicated(keep="last")
df[df["First Name"].duplicated(keep="last")]
```

```python
# it show value which is False by default keep is True

m=df["First Name"].duplicated(keep=False)
df.dropna(how='any',inplace=True)
df[m]
```
