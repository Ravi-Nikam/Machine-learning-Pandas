```python
# apply() in pandas
import pandas as pd
bond=pd.read_csv('jamesbond.csv')
bond.set_index("Film",inplace=True)
bond["Bond Actor Salary"].fillna(0,inplace=True)
bond.head()
```

```python
def add_mill(number):
    return str(number) + " Mill"
```

```python
bond["Box Office"]=bond["Box Office"].apply(add_mill)
bond.head()
```

```python
for col in ["Budget","Box Office","Bond Actor Salary"]:
    bond[col]=bond[col].apply(add_mill)
bond.head()    
```
