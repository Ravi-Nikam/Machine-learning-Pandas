```python
# logical operator AND and OR


import pandas as pd
df=pd.read_csv("employee.csv",parse_dates=['Start Date','Last Login Time']).dropna(how="all")
df["Team"].fillna("No Team",inplace=True)
df["Salary"].fillna(0,inplace=True)
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df
```

```python
# and operator
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "No Team"
mask3 = df["Salary"] > 120000
df[mask1 & mask2 & mask3]
```

```python
# or and not
mask1 = df["Bonus %"] < 1.5
mask2 = df["Team"] == "No Team"
mask3 = df["Start Date"] < "1996-03-31"
df[mask1 | mask2 & mask3]
```
