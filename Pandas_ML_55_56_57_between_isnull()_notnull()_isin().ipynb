```python
# isin() method in pandas

import pandas as pd
df=pd.read_csv("employee.csv", parse_dates=['Start Date','Last Login Time'])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df.info()
df.head()
```

```python
x=df["Team"].isin(["Legal","Product","Finance"])
df[x].head()
```

```python
df.head()
```

```python
# showing all the NaN values using isnull() method
df[df["Team"].isnull()] 
```

```python
# showing all the values which is not NaN
df[df["Team"].notnull()].head()
```

```python
# between Salary 60k to 70k


df[df["Salary"].between(60000,70000)].head()
```

```python
# between Date
df[df["Start Date"].between("1995-04-22","2015-06-28")].head()
```

```python
# between last login Time on other way

df[df["Last Login Time"].between("8:30AM","9:00AM")].head()
```
