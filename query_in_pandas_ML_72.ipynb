```python
# query helpful for filtering dataframe 
# Note query() only help us when column name can't contain space 
import pandas as pd
bond=pd.read_csv('jamesbond.csv')
bond.set_index("Film",inplace=True)
bond["Bond Actor Salary"].fillna(0,inplace=True)
bond.head()
```

```python
bond.columns=[column_name.replace(" ","_")for column_name in bond.columns]
bond.head()    
```

```python
#filter actor name as Sean Connery
bond.query("Actor == 'Sean Connery'")
```

```python
# filtering based on multiple conditon
bond.query("Actor =='Sean Connery' and Director =='Guy Hamilton'")
```

```python
# for fetching data from multiple operator
bond.query("Box_Office > 800 and Actor == 'Sean Connery'")
```

```python
# use of in operator
bond.query("Actor in ['George Lazenby','Sean Connery']")
```
