```python
import pandas as pd
df=pd.read_csv("employee.csv", parse_dates=['Start Date','Last Login Time'])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df.sort_values(["First Name"],inplace=True)
df.head()
```

```python
# use of drop duplicated to remove all the duplicate value on the basis of First Name

df.drop_duplicates(subset=["First Name"],keep="first")
```

```python
df.dropna(inplace=True)
df.drop_duplicates(subset=["First Name","Team"],keep="first",inplace=True)
df.head(10)
```
