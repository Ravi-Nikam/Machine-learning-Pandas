```python
# sample
import pandas as pd
bond=pd.read_csv('jamesbond.csv')
bond.set_index("Film",inplace=True)
bond.sort_index(inplace=True)
bond.head()
```

```python
# by default it show 1 row
bond.sample()
```

```python
bond.sample( n=5,
    frac=None,
    replace=False,
    weights=None,
    random_state=None,
    axis=None)
```

```python
# getting random column
bond.sample(n=1,
    frac=None,
    replace=False,
    weights=None,
    random_state=None,
    axis=1)
```

```python
# getting .25 means it show 25% dataset using frac(fraction)
bond.sample(n=None,
    frac=.25,
    replace=False,
    weights=None,
    random_state=None,
    axis=None)
```
