```python
from pandas import DataFrame
Cabin = [['T',0], ['M',1], ['A',1], ['G',1], ['C',0], ['F',0], ['B',0], ['E',1], ['D',0],['A',1],['D',0], ['B',0]]
data=DataFrame(Cabin,columns=['Cabin','Survived'])
data.head(10)
```

```python
# grouby cabin based on Survived column
data.groupby(['Cabin'])['Survived'].count()
```
