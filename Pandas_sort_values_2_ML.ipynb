```python
# sort_values

import pandas as pd

nba=pd.read_csv("nba.csv")
nba
```

```python

# sort value by Name
nba.sort_values(by="Name",ascending=False).head()
```

```python
nba.sort_values(by="Age",ascending=False).head()
```

```python
# NaN values last and first
nba.sort_values(by="Salary",na_position='last')
nba.sort_values(by="Salary",na_position='first')
```
