```python
# .where()
import pandas as pd
bond=pd.read_csv('jamesbond.csv')
bond.set_index("Film",inplace=True)
bond.sort_index(inplace=True)
bond.head()
```

```python
# show all values whose actor is Sean Connery and expect values are shown as NaN
mask = bond["Actor"] == "Sean Connery"
bond.where(mask)
```

```python
mask2 = bond["Box Office"] > 100
bond.where(mask & mask2)
```
